## 범죄현황 데이터 전처리

```python
crime = pd.read_csv("./data/서울시_범죄발생현황.csv")
crime
```

```python
crime = crime.drop(["자치구별(1)"], axis=1)
```

```python
crime = crime.rename(mapper={"자치구별(2)": "county_name", "2022": "total_crime", "2022.1": "total_arrest", 
                     "2022.2": "murder_crime", "2022.3": "murder_arrest",
                     "2022.4": "robbery_crime", "2022.5": "robbery_arrest", 
                     "2022.6": "sex_crime", "2022.7": "sex_arrest",
                     "2022.8": "theft_crime", "2022.9": "theft_arrest",
                     "2022.10": "violence_crime", "2022.11": "violence_arrest"}, axis=1)
crime
```

```python
crime.head(7)
```

```python
crime = crime.iloc[4:, :]
```

```python
crime = crime.reset_index()
```

```python
crime = crime.drop(["index"], axis=1)
```

```python
crime.columns
```

```python
crime["city_name"] = "서울"
crime = crime[["city_name", "county_name", 'total_crime', 'total_arrest', 'murder_crime',
       'murder_arrest', 'robbery_crime', 'robbery_arrest', 'sex_crime',
       'sex_arrest', 'theft_crime', 'theft_arrest', 'violence_crime',
       'violence_arrest']]
```

```python
crime.to_csv("./data3/crime.csv", index=False)
```
