```python
from google.oauth2 import service_account

credentials = service_account.Credentials.from_service_account_file("./data/key.json")
```

```python
from google.cloud import bigquery

client = bigquery.Client(
    credentials=credentials,
    project=credentials.project_id
)
```

```python
light_area = '''
SELECT 
  light.county_name,
  light.town_name,
  area.Subtotal,
  sum(light.install_num) as num_cnt
FROM 
  `my-fantazip.fantazip.town_area_final` as area
inner join
  `my-fantazip.fantazip.light` as light
on
  area.county_name = light.county_name and
  area.town_name = light.town_name
group by
  light.county_name,
  light.town_name,
  area.Subtotal
order by
  light.county_name
'''
```

```python
light_area = client.query(light_area).to_dataframe()
light_area
```

```python
light_area.columns
```

```python
light_area
```

```python
light_area['ratio'] = light_area['num_cnt'] / light_area['Subtotal']
county_ratio_sum = light_area.groupby('county_name')['ratio'].sum().sort_values(ascending=False)

# 바 그래프로 시각화합니다.
county_ratio_sum.plot(kind='bar')
plt.xlabel('County Name')
plt.ylabel('Mean Ratio (num_cnt / Subtotal)')
plt.title('Mean Ratio by County')
plt.xticks(rotation=90)
plt.show()
```

```python
# "num_cnt"와 "Subtotal"을 나누어 새로운 열 "ratio"를 만듭니다.
light_area['ratio'] = light_area['num_cnt'] / light_area['Subtotal']

# plt.figure() 내에서 figsize 옵션을 설정하여 그래프의 크기를 조정합니다.
light_area[light_area["county_name"]=="종로구"][["town_name", "ratio"]].plot(x="town_name", y="ratio", kind="bar")
```

```python
# top10
light_area[light_area["county_name"]=="종로구"][["town_name", "ratio"]].sort_values(by="ratio")[:10]
```

```python
# top10
light_area[light_area["county_name"]=="종로구"][["town_name", "ratio"]].sort_values(by="ratio")[-10:]
```

```python
light_area['county_name'].unique()
```

```python
# "county_name"별로 서브플롯을 생성하여 각각의 "county_name"에 대한 "town_name"별 "ratio"를 나타내는 바 그래프를 그립니다.
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(25, 25))
counties = light_area['county_name'].unique()

for i, county in enumerate(counties):
    ax = axes[i//5, i%5]
    county_data = light_area[light_area['county_name'] == county]
    county_data.plot(ax=ax, x='town_name', y='ratio', kind='bar', title=county, legend=False)
    ax.set_xlabel('Town Name')
    ax.set_ylabel('Ratio')
    ax.set_xticklabels(county_data['town_name'])
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
