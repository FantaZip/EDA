```python
from google.oauth2 import service_account

credentials = service_account.Credentials.from_service_account_file("./data/key.json")
```

```python
from google.cloud import bigquery

client = bigquery.Client(
    credentials=credentials,
    project=credentials.project_id
)
```

```python
police_pop = '''
SELECT 
  police.county_name,
  police.town_name,
  sum(pop.total) as total,
  count(police.department_name) as cnt
FROM 
  `my-fantazip.fantazip.police` as police
inner join
  `my-fantazip.fantazip.population_by_age` as pop
on
  police.county_name = pop.county_name and
  police.town_name = pop.town_name
group by
  police.county_name,
  police.town_name
'''
```

```python
police_pop = client.query(police_pop).to_dataframe()
police_pop
```

```python
police_pop["ratio"] = police_pop["cnt"] / police_pop["total"]
police_pop
```

```python
police_pop["county_name"].unique()
```

```python
police_pop.groupby("county_name")["cnt"].sum().plot(kind="bar")
```

```python
police_pop.groupby("county_name")["ratio"].sum().plot(kind="bar")
```

```python
police_pop[police_pop["county_name"]=="중구"][["town_name", "ratio"]].plot(kind="bar", x="town_name")
```

```python
fig, axes = plt.subplots(5, 5, figsize=(20, 20))
county_names = police_pop["county_name"].unique()

# "county_name"을 변경하면서 각각의 서브플롯에 그래프를 그립니다.
for i, county_name in enumerate(county_names):
    ax = axes[i // 5, i % 5]
    data = police_pop[police_pop["county_name"] == county_name][["town_name", "ratio"]]
    data.plot(kind="bar", x="town_name", ax=ax, legend=False)
    ax.set_title(county_name)
    ax.set_xlabel("Town Name")
    ax.set_ylabel("Ratio")
    ax.set_xticklabels(data["town_name"], rotation=90)

# 빈 서브플롯 숨기기
for i in range(len(county_names), 25):
    fig.delaxes(axes[i // 5, i % 5])

plt.tight_layout()
plt.show()
```
